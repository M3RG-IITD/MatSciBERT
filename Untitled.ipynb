```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from new_plot import *
set_things()
```

```python
root = os.getcwd()
ner_path=os.path.join(root,'results','extracted_entities')
```

```python
apl_path = os.path.join(ner_path,'APL_normalised.csv')
df_apl = pd.read_csv(apl_path, header=None, names=['entity','counts'])
#sanity check
df_apl = df_apl.sort_values(by='counts',ascending=False)
df_apl
```

```python
xapl = df_apl.entity[:10]
yapl = df_apl.counts[:10]
plt.bar(xapl,yapl)
plt.xticks(rotation='90')
plt.xlabel('Entities')
plt.ylabel('Frequency')
save_path = os.path.join(root,'results','figures','apl.png')
plt.savefig(save_path,dpi=1200,bbox_inches='tight')
plt.show()
```
